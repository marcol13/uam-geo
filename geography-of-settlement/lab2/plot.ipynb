```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read the data from the CSV file
filename = './data/migration_indices.csv'

data = pd.read_csv(filename)
```

```python
# Show whole dataframe
data
```

```python
def draw_scatter_with_migration_groups(scatter_data, x, y, group):
    plt.figure(figsize=(10, 6))
    groups = scatter_data[group].unique()
    for g in groups:
        group_data = scatter_data[scatter_data[group] == g]
        plt.scatter(group_data[x], group_data[y], label=g, cmap='viridis')
    plt.legend(title='Powiat')
    plt.xlabel(x)
    plt.legend(groups, title='Powiat')
    plt.ylabel(y)
    plt.plot([0, 0], [-10, 10], color='black', linewidth=0.5)
    plt.plot([-10, 10], [0, 0], color='black', linewidth=0.5)
    plt.plot([-10, 10], [-10, 10], color='black', linewidth=0.5)
    plt.plot([-10, 10], [10, -10], color='black', linewidth=0.5)
    plt.xlim(-10,10)
    plt.ylim(-10,10)

    plt.text(-3, 1, 'H', fontsize=12)
    plt.text(-1, 3, 'A', fontsize=12)
    plt.text(1, 3, 'B', fontsize=12)
    plt.text(3, 1, 'C', fontsize=12)
    plt.text(3, -1, 'D', fontsize=12)
    plt.text(1, -3, 'E', fontsize=12)
    plt.text(-1, -3, 'F', fontsize=12)
    plt.text(-3, -1, 'G', fontsize=12)

    plt.xticks(range(-10, 11, 2))
    plt.yticks(range(-10, 11, 2))
    plt.xlabel('Współczynnik salda migracji')
    plt.ylabel('Współczynnik przyrostu naturalnego')
    plt.title(f'Przyrost rzeczywisty w województwie Opolskim w 2003 roku')
    plt.show()
```

```python
draw_scatter_with_migration_groups(data, 'migration_balance', 'population_growth', 'region')
```
